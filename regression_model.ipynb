```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Reset display options to default values
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
df = pd.read_csv('schools_stations.csv')

df
```

```python
# Create a function that will run a regression model with the x variables as inputs
def regression_modeling(x_vars):
    X = df[x_vars]
    y = df['User Rating Count']

    # Add a constant to the model
    X = sm.add_constant(X)

    # Create the model
    model = sm.OLS(y, X)

    # Fit the model
    results = model.fit()

    return results.summary()
```

```python
x_variables = ['Metro_Stations_Count', 'Bus_Stations_Count', 'Number of Nearby Colleges', 'Number of Nearby Schools']
print(regression_modeling(x_variables))
```

```python
x_variables = ['Number of Nearby Colleges', 'Number of Nearby Schools']
print(regression_modeling(x_variables))
```
